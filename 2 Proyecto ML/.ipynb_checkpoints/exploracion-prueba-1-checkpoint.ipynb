```python
import pandas as pd
import numpy as np
import missingno as msngo
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_csv("training_tweets.csv").drop(columns="Unnamed: 0")
df.head()
```

```python
msngo.matrix(df);
```

```python
# Palabras positivas
positive_words = ['happiness','love', 'surprise', 'fun', 'relief', 'enthusiasm']

# Palabras negativas
negative_words = ['worry', 'sadness', 'hate', 'empty', 'boredom', 'anger']
```

```python
df["polaridad"] = np.where(
    df["sentiment"].isin(positive_words),
    "positivo",
    np.where(
        df["sentiment"].isin(negative_words),
        "negativo",
        np.nan # En este caso estos registros se eliminarían. Otra opción es considerar este caso como "neutral" (otra categoría de sentimiento)
    )
)

df["polaridad"].value_counts("%", dropna=False) * 100
```
