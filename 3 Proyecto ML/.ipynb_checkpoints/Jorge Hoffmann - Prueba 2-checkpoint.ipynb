## Feature Enineering


### Tratamiento vector objetivo
En esta secci√≥n se recodificar√°n ambos vectores objetivos.

````python
# Generamos el segundo vector objetivo
violencia = df_proc.loc[:, "pf_hands":"pf_other"]
condiciones = [ # Lista de arrays. Uno por cada columna de violencia
    df_proc[v] == 'Y' # Hubo violencia en la columna v.
    for v in violencia.columns
]

df_proc['has_violence'] = np.where(np.logical_or.reduce(condiciones), 'Y', 'N')

###### Nota al margen para entender la l√≥gica tras el logical_or.reduce #######

# Condiciones es una lista de arrays. Cada array es una columna de 'Violencia'
# np.logical_or toma dos argumentos (arrays) computa el or para cada posici√≥n entre los dos arreglos. 
# Nosotros queremos m√°s argumentos (arrays), por eso tomamos reduce
# Reduce evaluar√° iterativamente el or entre parejas de arrays que est√©n en la lista de condiciones

# np.multiply(2,3) = 6. Pero no podemos hacer np.multiply(2, 3, 6). 
# Entonces hacemos np.multiply.reduce([2, 3 , 6]) y eso multiplicar√° iterativamente de a pares y dar√° 36
````

````python
# Recodificamos el vector objetivo
df_proc['arstmade'] = np.where(df_proc['arstmade'] == 'Y', 1, 0)
df_proc['arstmade'].value_counts()
````

````python
df_proc['has_violence'] = np.where(df_proc['has_violence'] == 'Y', 1, 0)
df_proc['has_violence'].value_counts()
````

Finalmente, ahora que ambos vectores fueron recodificados, es posible ver su comportamiento.

````python
# Un gr√°fico por vector objetivo
plt.figure(figsize = (10,4))

plt.subplot(1, 2, 1)
sns.countplot(
    y=df_proc['arstmade'],                          
    order=df_proc['arstmade'].value_counts().index,
    palette="Blues_r")
plt.title('arstmade')

plt.subplot(1, 2, 2)
sns.countplot(
    y=df_proc['has_violence'],                          
    order=df_proc['has_violence'].value_counts().index,
    palette="Blues_r")
plt.title('has_violence')
plt.tight_layout()
````

**Comentarios**  
Como se dijo anteriormente, el vector objetivo de arrestos se encuentra considerablemente desbalanceado. Hay muy pocos casos con arstmade = 1 en el set. Por otra parte, tambi√©n hay un desbalance en has_violence, sin embargo, este es menor.

````python
# Importamos nuestros datos de test
df_2010 = pd.read_csv("2010_1perc.csv").drop(columns=["Unnamed: 0"])

# Notamos que en los datos del 2010 en vez de decir STATEN ISLAND (como el 2009) dice STATEN IS. 
df_2010['city'] = np.where(df_2010['city'] == 'STATEN IS', 'STATEN ISLAND', df_2010['city'])

# df_proc es nuestro DataFrame a trabajar
df_2010_proc, categoricas, numericas = create_suitable_dataframe_with_nan(df_2010)
````

````python
# Generamos el segundo vector objetivo (violencia)
violencia = df_2010_proc.loc[:, "pf_hands":"pf_other"]
condiciones = [ # Lista de arrays. Uno por cada columna de violencia
    df_2010_proc[v] == 'Y' # Hubo violencia en la columna v.
    for v in violencia.columns
]
df_2010_proc['has_violence'] = np.where(np.logical_or.reduce(condiciones), 'Y', 'N')

df_2010_proc['has_violence'] = np.where(df_2010_proc['has_violence'] == 'Y', 1, 0)
df_2010_proc['arstmade'] = np.where(df_2010_proc['arstmade'] == 'Y', 1, 0)
````

````python
# Eliminamos columnas por distintos motivos. Luego eliminamos los nulos
columnas_exceso_nulos = ['age', 'beat']
columnas_sin_varianza = ['comppct', 'linecm', 'compyear', 'year']
atributos_id = ["datestop", "addrpct", "repcmd", "revcmd", "dob", "timestop", "linecm", "detailcm", "ser_num", "perobs"]
columnas_violencia = list(violencia.columns)
columnas_eliminar = columnas_exceso_nulos + columnas_sin_varianza + atributos_id + columnas_violencia
df_2010_proc = df_2010_proc.drop(columns = columnas_eliminar)
df_2010_proc = df_2010_proc.dropna()
````

````python
# Recodificamos variables num√©ricas
variables_numericas_recodificar = ["month", "pct"]

for col in variables_numericas_recodificar:
    df_2010_proc[col] = col + df_2010_proc[col].astype(str) 
    
df_2010_dummies = pd.get_dummies(df_2010_proc, drop_first = False)

# Columna que nos falta pq s√≠ aparece en el 2009 y no en el 2010 (en el 2010 no existe SN como registro)
df_2010_dummies['haircolr_SN'] = 0
````

````python
# Generamos nuestros datos de test
X_test = df_2010_dummies.copy()
y_test_ar = X_test.pop('arstmade')
y_test_v = X_test.pop('has_violence')
````

# Probabilidad arresto
En esta secci√≥n calcularemos la probabilidad de arresto por barrio, sexo y raza. Cabe destacar que agruparemos distintas razas para que el resultado sea m√°s analizable. Tambi√©n omitiremos los datos de g√©nero unknown pq representan muy poca proporci√≥n.

````python
# Calculo las probabilidades
tmp_pr = gb_model_ar.predict_proba(X_test) # Probabilidad de arresto
tmp_pr.shape # Tiene 2 columnas, la primera es arresto = 0 y la segunda es arresto = 1
````

````python
# Junto las probabilidades al c√°lculo original
df_prob = pd.concat([
    df_2010_proc[['city', 'race', 'sex']].reset_index(drop = True), # Nuestras observaciones
    pd.DataFrame(tmp_pr[:, 1], columns = ['prob_arresto']) #Selecciono la del √≠ndice 1 pq la segunda columna es la de la clase 1 (arresto = 1)    
], axis = 1)

df_prob.head()
````

````python
df_prob['race'].value_counts()
````

Dado que principalmente hay B, Q y W, agruparemos razas.

````python
# agrupo race ya que tiene muchos valores. Esto lo hacemos seg√∫n lo definido en el Excel de apoyo
razas = [
    'A',
    'B',
    'I',
    'P',
    'Q',
    'W',
    'Z',
    'U'
]
razas_agrupado = [
    'Other',
    'Black',
    'Other',
    'Black',
    'White',
    'White',
    'Other', 
    'Other'
]
df_prob['race'] = df_prob['race'].replace(razas, razas_agrupado)
df_prob['race'].value_counts()
````

Tenemos hecha nuestra agrupaci√≥n de razas. A continuaci√≥n, hacemos algo similar con el sexo.

````python
# Vemos el conteo para datos de sexo
df_prob['sex'].value_counts()
````

````python
# Mantenemos solo F y M
df_prob = df_prob[df_prob['sex'].isin(['F', 'M'])]
df_prob['sex'].value_counts()
````

````python
# Cambiamos a Female y a Male
df_prob['sex'] = df_prob['sex'].replace(['F', 'M'], ['Female', 'Male'])
df_prob['sex'].value_counts()
````

En este punto, solo tenemos Male y Female.

````python
# Obtenemos las probabilidades promedio agrupando por ciudad, raza y sexo
probs = df_prob.groupby([
    'city', 'race', 'sex'
]).agg({
    'prob_arresto' : 'mean'
}).reset_index().sort_values(by=[
    'prob_arresto' 
])
````

````python
# Generamos una columna que nos servir√° para el gr√°fico
probs['Concat'] = probs['city'] + ' - ' + probs['race'] + ' - ' +probs['sex']
````

````python
plt.figure(figsize = (10, 17))
plt.plot(probs['prob_arresto'], probs['Concat'], 'o')
plt.title('Probabilidad de arresto promedio por raza, sexo y barrio')
plt.xlabel('Probabilidad')
plt.ylabel('raza, sexo y barrio')
plt.axvline(
    probs['prob_arresto'].mean(),
    color = 'red',
    label = r'prob $\bar X:$' + f"{round(probs['prob_arresto'].mean(), 2)}"
);
````

Se observa qe la mayor probabilidad de arresto es para mujeres negras u other (no blancas), pr√°cticmente en todas las ciudades. 

Se observa una muy baja probabilidad de arresto en mujeres de Staten Island, bajo el 0.02.  

Se observa que hay 11 combinaciones posibles sobre el promedio, mientras que hay 19 bajo el promedio.  White aparece en 3 ocasiones sobre el promedio, mientras que en 7 aparece abajo. Es decir, en un 30% de los casos white qued√≥ bajo el promedio, mientras que para las otras razas fue un 36% (11/30). Lo anterior no es suficiente para decir que existe un sesgo, ya que la diferencia es muy poca y puede deberse a error del modelo o sencillamente algo propio de la muestra. 

Pese a lo anterior, y a que no es posible concluir la existencia de un sesgo, no deja de sorprender que hay combinaciones que tienen 8 veces m√°s probabilidad de terminar un arresto. Una mujer negra de Staten Island seguramente ve muy distinto una detenci√≥n respecto a una mujer negra de Bronx.
